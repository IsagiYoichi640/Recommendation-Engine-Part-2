```python
import pandas as pd
from IPython import display#show data in nice format
```

```python
movies=pd.read_csv("/content/movies.csv")
```

```python
movies['genres']=movies['genres'].str.split('|')#convert genres to list
```

```python
#recomendation function
def recommend(movie_title,n=10):
  #find movie
  movie=movies[movies['title'].str.contains(movie_title,case=False)]
  if movie.empty:
    #display all movies
    return pd.DataFrame(columns=['title','genres'])
  #extract genres from matched movies
  movie_genre=set(movie.iloc[0]['genres'])
  #calculate the similarity
  scores=[]
  for index,row in movies.iterrows():
    common=0
    for genre in row['genres']:
      if genre in movie_genre:
        common+=1
    scores.append(common)
  movies['score']=scores
  #arrange the movies according to similarity
  result=movies[movies['title']!=movie.iloc[0]['title']]\
  .sort_values('score', ascending=False)\
  .head(n)
  #output of function
  return result[['title','genres']].reset_index(drop=True)
```

```python
#user input
title=input("Enter a movie title:")
recommend(title)
```
